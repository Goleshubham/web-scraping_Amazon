```python
import requests
from bs4 import BeautifulSoup
import json

urls = [
    "https://www.amazon.com/dp/B0C7VBWHLQ/",
    "https://www.amazon.com/dp/B0BCH68K6M/",
    "https://www.amazon.com/dp/B0C9VWK5H5/",
    "https://www.amazon.com/dp/B0C1J4MPBB/"
]

data_list = []

for url in urls:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    try:
        full_name = soup.find("span", attrs={"id": "productTitle"}).get_text(strip=True)
        name = full_name.split(" - ")[0]  
    except AttributeError:
        name = "Not available"

    try:
        model_number = soup.find("span", attrs={"class": "productDetailsText"}).get_text(strip=True)
    except AttributeError:
        model_number = "Not available"

    try:
        rating = soup.find("span", attrs={"class": "a-icon-alt"}).get_text(strip=True)
    except AttributeError:
        rating = "Not available"

    try:
        no_of_ratings = soup.find("span", attrs={"id": "acrCustomerReviewText"}).get_text(strip=True)
    except AttributeError:
        no_of_ratings = "Not available"

    try:
        price = soup.find("span", attrs={"class": "a-price-whole"}).get_text(strip=True)
    except AttributeError:
        price = "Not available"

    try:
        screen_size = soup.find("span", text="Screen Size").find_next("span").get_text(strip=True)
    except AttributeError:
        screen_size = "Not available"

    try:
        brand = soup.find("a", attrs={"id": "bylineInfo"}).get_text(strip=True)
    except AttributeError:
        brand = "Not available"

    
    try:
        image_url = soup.find("img", id="landingImage")["src"]
    except (AttributeError, TypeError, KeyError):
        image_url = "Not available"

    
    additional_information = {
        "Brand Name": brand,
        "Screen Size": screen_size,
    }

    product_data = {
        "Name": name,
        "Model Number": model_number,
        "Rating": rating,
        "No of Rating": no_of_ratings,
        "Price": price,
        "Image URL": image_url,
        "Additional Information": additional_information,
        "URL": url,
    }

    
    details_table = soup.find("table", class_="a-keyvalue prodDetTable")
    if details_table:
        rows = details_table.find_all("tr")
        for row in rows:
            try:
                key = row.find("th").get_text(strip=True)
                value = row.find("td").get_text(strip=True).replace("\u200e", "")
                additional_information[key] = value
            except AttributeError:
                continue

    data_list.append(product_data)

print(json.dumps(data_list, indent=4))
```
